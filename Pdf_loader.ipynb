```python
# Lazy Loader-->Lazy load() Waits until you explicitly call .load() to instantiate and perform I/O.
# it loads on demand
# not loaded at once they are fetched one at a time as needed.
# Best to use when many files or larger documents
# retuns a generator of document
#load is caled eager loading(not advisable fo larger documents)
```

```python
documents_lazy=loader.lazy_load()

for doc_lazy in documents_lazy:
    print(doc_lazy)
```
